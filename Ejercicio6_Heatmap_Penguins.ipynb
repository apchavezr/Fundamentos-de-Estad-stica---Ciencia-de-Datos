# Ejercicio 6: Heatmap de correlaciones en el conjunto `penguins`

Este ejercicio tiene como objetivo construir un **mapa de calor (heatmap)** que muestre las correlaciones entre las variables numéricas del conjunto de datos `penguins`, disponible en la librería Seaborn.

A partir del análisis visual de la matriz de correlación, se busca identificar **qué variables están más fuertemente relacionadas** y discutir las posibles implicaciones para un análisis predictivo.

---
## 1. Cargar el conjunto de datos


```python
import seaborn as sns
import pandas as pd

# Cargar el conjunto de datos
df = sns.load_dataset("penguins")

# Mostrar los primeros registros
df.head()
```

---

## 2. Construcción del mapa de calor de correlaciones

Se consideran únicamente las variables numéricas para calcular la matriz de correlación.


```python
import matplotlib.pyplot as plt

# Filtrar solo columnas numéricas
df_num = df.select_dtypes(include='number')

# Calcular la matriz de correlación
correlation_matrix = df_num.corr()

# Visualizar el heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Mapa de correlaciones entre variables numéricas")
plt.show()
```

---

## ✅ 3. Análisis de la correlación

**Observaciones destacadas:**

- Existe una **correlación positiva fuerte** entre `flipper_length_mm` y `body_mass_g` (mayor a 0.85), lo que indica que los pingüinos con aletas más largas tienden a tener mayor masa corporal.
- También se observa una correlación positiva entre `bill_length_mm` y `bill_depth_mm`, aunque de menor magnitud.
- La variable `island` y otras categóricas no fueron incluidas en este análisis por no ser numéricas.

**Implicaciones en el análisis predictivo:**

- Si se utiliza un modelo de regresión, es importante considerar el impacto de la **colinealidad** entre variables altamente correlacionadas, ya que pueden influir negativamente en la interpretación de los coeficientes.
- En algoritmos como árboles de decisión o bosques aleatorios, la alta correlación puede redundar en **importancia compartida** entre variables, sin afectar tanto el desempeño predictivo.
