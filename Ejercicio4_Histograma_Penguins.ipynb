# Ejercicio 1: Análisis exploratorio con histogramas y curva de densidad

Este ejercicio tiene como objetivo practicar técnicas de visualización exploratoria utilizando el conjunto de datos `penguins` de la librería Seaborn. Se construirá un histograma para la variable `body_mass_g` acompañado de una curva de densidad para analizar su distribución.

---
## 1. Cargar los datos


```python
import seaborn as sns
import pandas as pd

# Cargar el conjunto de datos penguins
df = sns.load_dataset("penguins")

# Mostrar las primeras filas
df.head()
```

---

## 2. Visualización de la variable `body_mass_g`

Se construirá un histograma y una curva de densidad para observar la distribución de la masa corporal de los pingüinos.


```python
import matplotlib.pyplot as plt

# Crear histograma con curva de densidad
plt.figure(figsize=(8, 5))
sns.histplot(df['body_mass_g'].dropna(), kde=True, color='skyblue', bins=20)
plt.title("Distribución de la variable body_mass_g")
plt.xlabel("Masa corporal (g)")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.show()
```

---

## ✅ 3. Análisis de la distribución

Al observar el histograma con la curva de densidad, se puede hacer el siguiente análisis:

- La distribución presenta **más de una moda**, lo cual indica **multimodalidad**. Esto sugiere que podrían existir subgrupos en los datos, posiblemente asociados a las distintas especies de pingüinos.
- La distribución no muestra un **sesgo fuerte**, aunque hay una ligera cola hacia la derecha.
- Es recomendable continuar la exploración separando por especie para confirmar si la multimodalidad está asociada a esa variable.

Este tipo de análisis es útil para decidir estrategias de modelamiento posteriores y para entender mejor la estructura de los datos.
