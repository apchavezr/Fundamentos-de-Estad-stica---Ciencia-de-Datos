# Ejercicio 5: Comparación de variables mediante boxplots

En este ejercicio se utilizarán diagramas de caja (boxplots) para comparar la longitud de aletas (`flipper_length_mm`) y la longitud del pico (`bill_length_mm`) entre las distintas especies del conjunto de datos `penguins` de la librería Seaborn.

Se busca identificar qué especies presentan mayor dispersión en estas variables.

---
## 1. Cargar el conjunto de datos


```python
import seaborn as sns
import pandas as pd

# Cargar dataset
df = sns.load_dataset("penguins")

# Mostrar primeras filas
df.head()
```

---

## 2. Boxplot de la variable `flipper_length_mm` por especie

Se graficará la longitud de la aleta según la especie.


```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x="species", y="flipper_length_mm", palette="Set2")
plt.title("Distribución de flipper_length_mm por especie")
plt.xlabel("Especie")
plt.ylabel("Longitud de aleta (mm)")
plt.grid(True)
plt.show()
```

---

## 3. Boxplot de la variable `bill_length_mm` por especie

Ahora se graficará la longitud del pico por especie.


```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x="species", y="bill_length_mm", palette="Set3")
plt.title("Distribución de bill_length_mm por especie")
plt.xlabel("Especie")
plt.ylabel("Longitud del pico (mm)")
plt.grid(True)
plt.show()
```

---

## ✅ 4. Análisis de la dispersión

**Observaciones:**

- Para `flipper_length_mm`, la especie **Gentoo** presenta mayor dispersión, seguida de cerca por **Adelie**.
- En el caso de `bill_length_mm`, se observa mayor variabilidad en la especie **Adelie**, con una mayor amplitud intercuartílica y presencia de outliers.
- Estas diferencias en dispersión pueden reflejar tanto la variabilidad morfológica dentro de cada especie como posibles errores de medición o sesgo en la recolección de datos.

El uso de boxplots permite identificar fácilmente estas características y facilita la comparación entre grupos.
