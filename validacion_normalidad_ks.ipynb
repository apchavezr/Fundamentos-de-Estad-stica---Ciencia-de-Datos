# Validaci√≥n de Supuestos: Normalidad con Kolmogorov-Smirnov

```python
import numpy as np
from scipy.stats import kstest, norm
import seaborn as sns
import matplotlib.pyplot as plt
```

### Paso 1: Generar residuos simulados de un modelo

```python
np.random.seed(42)
residuos = np.random.normal(loc=0, scale=1, size=100)
sns.histplot(residuos, kde=True)
plt.title('Distribuci√≥n de residuos')
plt.show()
```

### Paso 2: Aplicar prueba Kolmogorov-Smirnov

```python
ks_stat, p_valor = kstest(residuos, 'norm')
print(f"Estad√≠stico K-S: {ks_stat:.4f}")
print(f"p-valor: {p_valor:.4f}")
```

### Conclusi√≥n
Si el p-valor > 0.05, se acepta que los residuos siguen una distribuci√≥n normal.
