# Ejercicio: Prueba ANOVA con datos simulados

En este ejercicio se aplica una prueba ANOVA para evaluar si existen diferencias significativas en los promedios de tres grupos simulados (A, B y C).

```python
import pandas as pd
import numpy as np
from scipy import stats
```

### Paso 1: Cargar los datos

```python
df = pd.read_csv('datos_simulados_anova.csv')
df.head()
```

### Paso 2: Descripción general de los datos

```python
df.groupby('grupo')['valor'].describe()
```

### Paso 3: Aplicar prueba ANOVA de un solo factor

```python
grupo_A = df[df['grupo'] == 'A']['valor']
grupo_B = df[df['grupo'] == 'B']['valor']
grupo_C = df[df['grupo'] == 'C']['valor']

f_stat, p_valor = stats.f_oneway(grupo_A, grupo_B, grupo_C)
print(f"Estadístico F: {f_stat:.4f}")
print(f"p-valor: {p_valor:.4f}")
```

### Paso 4: Conclusión

Si el p-valor es menor que 0.05, se concluye que existen diferencias significativas entre al menos dos grupos.
