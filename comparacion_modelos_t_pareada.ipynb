# Comparación de Modelos Predictivos usando prueba t pareada

```python
import numpy as np
from scipy.stats import ttest_rel
import matplotlib.pyplot as plt
```

### Paso 1: Simulación de métricas de validación cruzada

```python
np.random.seed(42)
modelo_A = np.random.normal(loc=0.85, scale=0.02, size=10)
modelo_B = np.random.normal(loc=0.83, scale=0.02, size=10)
```

```python
plt.plot(modelo_A, label='Modelo A')
plt.plot(modelo_B, label='Modelo B')
plt.title('Precisión en 10 folds')
plt.legend()
plt.show()
```

### Paso 2: Aplicación de prueba t pareada

```python
t_stat, p_valor = ttest_rel(modelo_A, modelo_B)
print(f"Estadístico t: {t_stat:.4f}")
print(f"p-valor: {p_valor:.4f}")
```

### Conclusión
Si el p-valor es menor a 0.05, se concluye que uno de los modelos tiene un rendimiento significativamente mejor.
